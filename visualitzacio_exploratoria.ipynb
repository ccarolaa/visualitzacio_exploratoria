Crea almenys una visualització per:

Per a resoldre aquest exercici, optem per realizar visualitzacions gràfiques diverses utilitzant els dos paquest matplotlib i seaborn. L'objetiu és representar les diferents relacions entre camps del dataset "DelayedFLights" per poder extreure conclusions interessants a la vegada que es mostra una gran varietat de gràfics útils per l'anàlisi de dades.

````python
# Importem els paquets

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Carreguem el dataset

delays = pd.read_csv('C:/Users/34607/Desktop/DelayedFlights.csv',index_col='Unnamed: 0', parse_dates= {"Date" : ["Year","Month","DayofMonth"]})
````

````python
delays
````

````python
# Esborrem les columnes que no volem per a l'anàlisi

delays = delays.drop(['FlightNum','TailNum','TaxiIn','TaxiOut','CancellationCode'], axis =1)
````

**Visualització d'una variable categòrica (UniqueCarrier)**

````python
# Creem un histograma mitjançant la funció countplot() del paquet seaborn, que mostra el recompte de registres per a cada categoria (en aquest cas, cada aerolinia).

sns.countplot(x='UniqueCarrier',data=delays)

# Posem un títol al gràfic i als dos eixos:
plt.title('UniqueCarrier')
plt.xlabel("Aerolínies")
plt.ylabel("Número de vols")

# Per a guardar els gràfics mostrats, utilizem la funció savefig():
plt.savefig('plot1.png')

# Per a mostrar el gràfic, utilitzem la funció show():
plt.show()
````

Podem observar que la aerolínia amb més vols al dataset es 'WN' amb més de 350.000 registres, seguida de 'AA' amb 200.000 registres aproximats.

**Visualització d'una variable numèrica (ArrDelay)**

````python
# El següent gràfic es crea amb la funció plot(). Es visualitza la suma de minuts de retràs dels avions a l'arribada al aeroport de día, per dia.

plt.plot(delays.groupby('Date')['ArrDelay'].sum(), color='blueviolet',linewidth=0.5)
plt.title('Suma de minuts de retràs a l´arribada per dia')
plt.xlabel("Dia")
plt.ylabel("Suma de minuts de retràs a l'arribada")
plt.savefig('plot2.png')
plt.show()
````

Algunes conclusions del gràfic anterior són que els dies amb més minuts de retràs es situan a finals de l'any 2008 i que de setembre a novembre, la variació entre dies dels minuts de retràs és estable i més baixa que a la resta de l'any.

**Visualització d'una variable numèrica i una categòrica (ArrDelay i UniqueCarrier)**

````python
# En aquest cas, utilitzem la funció catplot() del paquet seaborn, idonea per mostrar la relació entre una variable numèrica i una categòrica. Cada punt representa un registre de minuts de retràs a l'arribada, per a cada aerolínia.

sns.catplot(x='UniqueCarrier',y='ArrDelay',data=delays)
plt.title('Minuts de retràs a l´arribada per aerolínia')
plt.xlabel("Aerolínies")
plt.ylabel("Minuts de retràs a l´arribada")
plt.savefig('plot3.png')
plt.show()
````

Podem observar que totes les aerolínies presenten minuts de retrasos a l'arribada i sobretot destaca per la aerolinia NW, MQ i AA (en aquest ordre), que presenten el major número de vols amb més minuts de retràs a l'arribada.

**Visualització de dues variables numèriques (ArrDelay i DepDelay)**

````python
# Creem unes mostres del dataset per a reduir els temps d'execució de cada funció dels següents exercicis.

s_delays = delays.sample(10000)
s_delays
s1_delays = delays[delays['UniqueCarrier'].isin(['AA','NW','B6','OO'])].sample(5000)
print(s1_delays)
````

````python
# S'utilitza la funció scatterplot() del paquet seaborn, on cada punt és un registre que té com a valor d'ArrDelay el valor de l'eix X, i com a valor de DepDelay el valor de l'eix y.

sns.scatterplot(data=s_delays, x="ArrDelay", y="DepDelay",color="limegreen")
plt.title('Minuts de retràs a la sortida i a l´arribada del vol')
plt.xlabel("Minuts de retràs a l'arribada")
plt.ylabel("Minuts de retràs a la sortida")
plt.savefig('plot4.png')
plt.show()
````

Alguns punts que podem destacar del gràfic anterior són:
* La relació entre ArrDelay i DepDelay sembla positivament correlacionada.
* Podem veure que, per norma general, si hi ha X minuts de retràs a la sortida, hi ha els mateixos X minuts o més minuts de retràs a l'arribada.

**Visualització de tres variables (ArrDelay, DepDelay i UniqueCarrier)**

````python
# En aquest cas, utilitzem la mostra de 5.000 vols per a 4 aerolínies: AA, B6, OO, NW. Executem la mateixa funció scatterplot(), afegint el paràmetre "hue" amb la variable categòrica UniqueCarrier.

sns.scatterplot(x='ArrDelay', y='DepDelay', data=s1_delays, hue='UniqueCarrier',palette='Set2')
plt.title('Minuts de retràs a la sortida i a l´arribada, per aerolínia')
plt.xlabel("Minuts de retràs a l'arribada")
plt.ylabel("Minuts de retràs a la sortida")
plt.savefig('plot5.png')
plt.show()
````

Podem veure que amb el paràmetre "hue", es pot especificar la companyia aeria de cada vol mitjançant el codi de colors i aerolinies de la llegenda. A la cantonada inferior esquerra, podem observar una agrupació de registres on es barrejen totes les aerolínies. No obstant, hi ha certes observacions que presenten una relació de minuts de retràs a la sortida i a la arribada superiors a la resta. Estan a la cantonada superior dreta i són de les aerolínies NW (en rosa) i AA (en verd).

**Visualització de més de tres variables (ArrDelay, DepDelay, AirTime i UniqueCarrier)**

````python
# Per la última visualització, optem per un "boxplot", on es pot observar la forma de la distibució i on es situen la mediana i els valors centrals de cada columna. 

sns.catplot(x='UniqueCarrier',y='ArrDelay',data=s1_delays,palette='Set2',kind = 'box')
plt.title('Minuts de retràs a l´arribada per aerolínia')
plt.xlabel("Aerolínies")
plt.ylabel("Minuts de retràs a l´arribada")
plt.savefig('plot6.png')
sns.catplot(x='UniqueCarrier',y='DepDelay',data=s1_delays,palette='Set2',kind = 'box')
plt.title('Minuts de retràs a la sortida per aerolínia')
plt.xlabel("Aerolínies")
plt.ylabel("Minuts de retràs a la sortida")
plt.savefig('plot7.png')
sns.catplot(x='UniqueCarrier',y='AirTime',data=s1_delays,palette='Set2',kind = 'box')
plt.title('Minuts de vol per aerolínia')
plt.xlabel("Aerolínies")
plt.ylabel("Minuts de vol")
plt.savefig('plot8.png')
plt.show()
````

Aquestes visualitzacions ens aporten molta informació sobre la relació de les tres variables ArrTime, ArrDelay i DepDelay i les aerolínies, pero també entre les tres variables entre elles.
Una d'aquestes conclusions és que la aerolinia B6 és la que presenta més minuts de retràs a l'arribada i a la sortida ja que el seu segon i quartil es situen a una quantitat de minuts de retràs superiors a la resta de companyies. No obstant, el seu valor màxim de minuts de retràs tant a l'arribada com a la sortida és inferior al màxim de la resta d'aerolínies, com podem veure als valors atípics superiors. Una altra observació és que el rang dels minuts de vol per aerolínia és inferior al rang de minuts de retràs a la sortida o a l'arribada, el que ens porta a pensar que els "problemes" que provoquen els retrasos es donen abans de que l'avio s'enlairi o desprès que l'avió arribi al destí.
